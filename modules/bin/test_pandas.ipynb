```python
output_list = []

with open("sample_book_text.txt", 'r') as f:
    for line in f:
        # Maybe use REGEX for this... Test in jupyter, and write separate function
        line = line.replace(".", "").replace(":", "").replace(",", "").replace("*", "")
        line = line.replace("_", "").replace("-", "").replace("(", "").replace("[", "")
        line = line.replace("\"", "")
        words = line.split()
        for word in words:
            word = word.strip().lower()
            output_list.append(word)

```

```python
def histogram_as_dict(source_text_as_list):
    """Return histogram as a dict: {'one': 1, 'fish': 4}..."""
    output_dict = {}

    for word in source_text_as_list:
        if word in output_dict.keys():
            output_dict[word] += 1
        else:
            output_dict[word] = 1

    return output_dict
```

```python
dict_output = histogram_as_dict(output_list)
```
